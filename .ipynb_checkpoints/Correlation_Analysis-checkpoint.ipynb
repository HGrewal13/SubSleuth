```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./shopping_behavior_updated.csv")
df.head()
```

```python
cleaned_df = df.drop(columns = {"Location", "Size", "Color", "Season", "Shipping Type"})
```

```python
cleaned_df.head()
```

```python
cleaned_df['Subscription Status Numeric'] = cleaned_df['Subscription Status'].map({'Yes': 1, 'No': 0})
```

```python
cleaned_df.head()
```

```python
correlation = cleaned_df[['Purchase Amount (USD)', 'Subscription Status Numeric']].corr()
print(correlation)
```

```python
cleaned_df['Frequency of Purchases'] = cleaned_df['Frequency of Purchases'].map({'Annually': 1, 'Quarterly': 2, 'Every 3 Months': 2, 'Monthly': 3, 'Bi-Weekly': 4, 'Fortnightly': 4, 'Weekly': 5})
```
